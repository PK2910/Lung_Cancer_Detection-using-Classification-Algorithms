```python
x=df[['Coughing of Blood','Passive Smoker','Obesity','Dust Allergy','Genetic Risk']]
y=df['Level']
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
xs=ss.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(xs,y,test_size=0.20,random_state=42)
print(x_train.shape)
print(y_test.shape)
```

```python
print(x_train)
print(x_test)
```

```python
from sklearn.metrics import classification_report
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
model=lr.fit(x_train,y_train)
model
```

```python
y_pred=model.predict(x_test)
print(model.coef_)
print(model.intercept_)
print(y_pred)
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.metrics import recall_score
from sklearn.metrics import precision_score
from sklearn.metrics import f1_score
```

```python
confusion_matrix(y_test,y_pred)
```

```python
accuracy_score(y_test, y_pred)
```

```python
recall_score(y_test, y_pred,average='macro')
```

```python
precision_score(y_test, y_pred,average='macro')
```

```python
f1_score(y_test, y_pred,average='macro')
```

```python
print(classification_report(y_test,y_pred))
```

# KNN

```python
#Cross Validation for choosing the best neighbours using GridSearchCV
from sklearn.neighbors import KNeighborsClassifier
```

```python
acc = []
for k in range(1, 39, 2):
    knn = KNeighborsClassifier(n_neighbors= k)
    sample_model=knn.fit(x_train,y_train)
    y1_pred = sample_model.predict(x_test)
    acc.append(accuracy_score(y_test, y1_pred))
print(acc)
```

```python
plt.figure(figsize = (12, 7))
plt.plot(range(1, 39, 2), acc)
plt.title("Acc vs K- Value")
print(f"maximum accuracy = {max(acc)} at k = {(np.array(acc).argmax())*2 + 1}")
plt.show()
```
